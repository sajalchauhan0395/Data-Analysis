Payment distribution among Males

```python
male_df = df1[df1['Gender'] == 'Male']
male_payment_counts = male_df['Payment'].value_counts() # Count the occurrences of each payment mode among males
male_payment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title('Payment Modes Distribution in Males')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Payment distribution among females


```python
female_df = df1[df1['Gender'] == 'Female']
female_payment_counts = female_df['Payment'].value_counts() # Count the occurrences of each payment mode among females
female_payment_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title('Payment Modes Distribution among females')

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
